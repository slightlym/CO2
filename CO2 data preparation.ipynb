Now the same with the file on the consumption emissions.

```python
cons = pd.read_csv('Consumption emissions.csv')
```

```python
cons.head(15)
```

```python
cons.iloc[0,0] = 'YYYY'
cons.head()
```

```python
cons_18 = cons_18.transpose().reset_index()
cons_18.head()
```

```python
cons_18.columns = ['country','consumption_emissions']
cons_18 = cons_18.iloc[1:]
cons_18.head()
```

```python
cons_18.dtypes
```

```python
cons_18 = cons_18.astype({'consumption_emissions': np.float})
```

Preparing the file with per capita data.

```python
per_capita = pd.read_csv('GCB2020v18_MtCO2_flat.csv', encoding = 'unicode_escape', engine = 'python')
```

```python
per_capita.head()
```

```python
per_capita.dtypes
```

```python
per_capita_18 = per_capita.loc[per_capita.Year == 2018,['Country','Total','Per Capita']]
per_capita_18.head()
```

```python
per_capita_18.loc[per_capita_18.Country == 'Russia']
```

```python
per_capita_18.Country.replace('Russia', 'Russian Federation', inplace = True)
per_capita_18.head()
```

Merge the data

```python
display(terr_18.shape, cons_18.shape, per_capita_18.shape)
```

```python
df = terr_18.merge(cons_18, on = 'country', how = 'inner').merge(per_capita_18, left_on = 'country', right_on = 'Country', how = 'inner')
```

```python
df.head()
```

```python
df.shape
```

Units of measurement are different in different initial files.

`territory_emissions` and `consumption_emissions` are in million tonnes of carbon (MtC), while `Total` from `per_capita` file is in million tonnes of CO2 (MtCO2) and `Per Capita` is in tCO2/person. There are 3.664 MtCO2 in 1 MtC.

Convert into `territory_emissions` and `consumption_emissions` into MtCO2.

```python
df['territory_MtCO2'] = df.territory_emissions * 3.664
df['consumption_MtCO2'] = df.consumption_emissions * 3.664
df['check'] = df.territory_MtCO2 - df.Total
df.check.describe()
```

```python
df.sort_values(by = 'check').head()
```

Only for France there is a significant difference between `territory` and `per_capita`. We ignore it assuming that overall there are no contradictions between two sources of data.

Calculate population size (in millions).

```python
df['population'] = df.Total / df['Per Capita']
df.head()
```

Calculate consumption emissions per capita.

```python
df['consumption_per_capita'] = df.consumption_MtCO2 / df.population
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.shape
```

Removal of missing data leaves us with roughly half of countries. Still, it is enough for the analysis.

```python
df.to_csv('CO2_prepared.csv')
```
